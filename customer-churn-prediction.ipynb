```python
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df.shape
```

```python
df = pd.get_dummies(df, columns=['Geography','Gender'], drop_first=True, dtype=int)
```

```python
df
```

```python
# let us scale the values first before entering it into neural network
```

```python
X = df.drop(columns=['Exited'])
y = df['Exited']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_train_scaled
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
#using sequential model
model = Sequential()
model.add(Dense(11,activation='relu',input_dim = 11)) #in our hidden layer we are putting 3 perceptrons which would be getting 11 inputs
model.add(Dense(11,activation='relu'))
model.add(Dense(1,activation='sigmoid')) # one output
```

```python
model.summary()
```

```python
model.compile(
    loss='binary_crossentropy',
    optimizer='Adam',
    metrics=[
        'accuracy',
        tensorflow.keras.metrics.Precision(),
        tensorflow.keras.metrics.Recall(),
        tensorflow.keras.metrics.AUC()
    ],
    
)
```

```python
model.fit(X_train_scaled,y_train,epochs=100,validation_split=0.2)
```

```python
#now let us see the values of teh calculated weights and biases
model.layers[0].get_weights()
```

```python
model.layers[1].get_weights()
```

```python
model.layers[1].get_weights()
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python

model.predict(X_test_scaled)
```

```python
# now these are teh sigmoid probabilities , we now have to decide a threshold , above which there will be one and below which there will be zero
y_log = model.predict(X_test_scaled)
```

```python
y_pred = np.where(y_log>0.5,1,0)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
